```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('netflix_titles.csv')  
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
# See actual column names first
print(df.columns)

# Then try filling with correct names
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Not Rated', inplace=True)
```

```python
print(df.columns)

# Fill missing director values with 'No Director Info'
df['director'].fillna('No Director Info', inplace=True)

# Fill missing cast values with 'No Cast Info'
df['cast'].fillna('No Cast Info', inplace=True)
```

```python
df[['director', 'cast']].isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['type'] = df['type'].str.strip().str.lower()
df['country'] = df['country'].str.strip().str.title()
```

```python
print(df.columns)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df['release_year'] = df['release_year'].astype(int)
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df.info()
df.isnull().sum()
df.duplicated().sum()
```

```python
df.to_csv('cleaned_netflix_data.csv', index=False)
```
